```python
# Create a histogram based on generated data
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Generate random data
data = np.random.randn(1000)
df = pd.DataFrame(data, columns=['Value'])

# Set the style
sns.set(style="whitegrid")

# Create the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Value'], bins=30, kde=True, color='blue', stat='density')  
plt.title('Histogram of Generated Data with KDE')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
